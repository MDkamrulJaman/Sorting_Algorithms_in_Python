# Bubble Sort


# Why n - i - 1:

*   n: Total number of elements in the list.

*   i: The current iteration of the outer loop.
*   n - i - 1: Limits the number of inner comparisons because after each pass, the largest unsorted element has "bubbled" to its correct position.

* n - i - 1: Limits the number of inner comparisons because after each pass, the largest unsorted element has "bubbled" to its correct position.

### range(start, stop)
Here, start defaults to 0, so it's equivalent to range(0, n - i - 1).

### To optimize and stop unnecessary passes  to check if any swaps occurred during a pass. If no swaps are made in a pass, it means the list is already sorted, and the algorithm can exit early.




```python
# check array is already sorted or not
def is_sorted(arr):
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            return False
    return True
```

```python


def bubble_sort(arr):

    # Length of the array
    n = len(arr)

    outer_loop_count = 0
    inner_loop_count = 0

    # Outer loop for access each array element
    for i in range(n-1):
     outer_loop_count += 1

     if is_sorted(arr):
            print("The array is already sorted, no iterations needed.")
            break


     else:
        # Show current pass number
        print(f"Pass {i + 1}:")



        # Inner loop for comparing array element
        for j in range(0, n - i - 1):

            inner_loop_count += 1

            if arr[j] > arr[j + 1]:  # Comparing adjacent elements

                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # swapping elements

                print(f"swap {arr[j+1]} <-> {arr[j]} : {arr}")




    print(f"Outer loop ran {outer_loop_count} times")
    print(f"Inner loop ran {inner_loop_count} times")

    return arr



```

```python

# input function from the user
def take_input():
    # while True:
        unsorted_list = input("Enter digits without spaces: ")

        if unsorted_list.isdigit():

            return [int(digit) for digit in unsorted_list] # string to a list

        else:
            print("Invalid input. Please enter only digits.")

```

```python
# output function

def display_output(unsorted_list):

    print("list:", unsorted_list)

    sorted_list = bubble_sort(unsorted_list)
    print("Sorted list:", sorted_list)
```
